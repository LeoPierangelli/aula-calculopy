```python
sy.plot(-4*x+3, (x, -1,1), legend = True);
```

```python
sy.plot(-4*x+3, 3*x+1, (x, -1,1), legend = True);
```

```python
sy.plot(4*x-1, (x, -0.5,1), legend = True)
```

## Questão 1

### Uma pequena empresa que fabrica blusas tem um custo fixo mensal de 4000,00. O custo de produção de uma blusa é de 20,00 e cada blusa é vendida por 70,00. Considere que todas as x unidades produzidas mensalmente são vendidas.

#### Determine a função que represemta o lucro mensal desta empresa

```python
def lucro(x):
  return 50*x - 4000
```

#### Construa o gráfico dessa função

```python
x = np.linspace(0,100)
plt.plot(x, lucro(x));
plt.grid()
plt.show()
```

#### Qual o lucro da empresa num mês que foram vendidas 500 blusas?

```python
lucro(500)
```

# Segundo Grau

```python
def g(x):
  return x**2 - 2*x -3
```

```python
def h(x):
  return -x**2 - 4
```

```python
sy.solve(g(x), x)
```

```python
sy.plot(g(x), legend = True);
```

```python
sy.plot(h(x), legend = True);
```
